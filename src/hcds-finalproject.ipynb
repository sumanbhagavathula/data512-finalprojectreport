# Exploring whether colder temperatures affect the number of people who commute by bike in seattle

## Introduction

In this work, I would like to explore through data whether the colder winter temperatures have any impact on the number of people who commute by bike in Seattle. The primary motivation for this is that I am a very keen bike rider but since I am originally from a place in South India where temperatures barely go into teens in the centigrade scale. In contrast, the winter temperatures here in Seattle are many a time so frigid for me that I do not even venture to stay out much, forget about getting to take my bike out. But that may just be me. Or is it? With Seattle being a cosmopoliton city and having attracted so many people from many geographic regions with varying weather conditions, I may not be lonely. If that it is the case, then there is a good chance that there could be like minded people interested in coming up with ideas and collectively implement them for the benefit of the group. 
A secondary benefit of large bike commuters is that it is an alternative means of transport, which means if all or many of these riders take a motorized transport option then there would be higher volumes of motor vehicles on the road which in turn also contributes to increase in emissions (to the extent where the vehicles are non-zero emission categories). 


## Research Overview

To explore whether the colder temperatures affect the number of people commuting by bike in the City of Seattle, we need to know how many people commute by bike in Seattle and also what the temperatures are at that place during the times the bike count measurements are taken. Fortunately, the Seattle Department of Transporation has both of these information for various locations throughout the city and have made these datasets publicly available. The details of these datasets will be presented in the Dataset overview section below. For now, I would like to introduce two location points in the City of Seattle where I wish to explore the bike commuter counts as temperatures change across days of the year. The first reference point is towards the south end of the City of Seattle and is called Mountain to Sound Trail West of I90. This is the point where the I90 touches the city of Seattle. The second reference point is towards the north end of the city of Seattle and is called the Burke Gilman Trail North of 70th ST. 

## Research Questions and Hypothesis

With the above background, below were my research questions followed by the hypothesis:

* RQ1: Taking a reference point Mountain to Sound Trail West of I90, which is close to where I90 touches the city of Seattle, do the number of bike commuters at this point reduce in number on the colder days of winter compared to the warmer days during the rest of the year?

* RQ2: Considering another reference point Burke Gilman Trail North of 70th ST, is the trend similar to what is observed at the first reference point Mount to Sound Trail West of I90?

For the first research question RQ1, I had the following hypothesis:

* H1: On any given day, there are at least some people who commute to work by bike in Seattle.
* H2: There are at least some days in winter in Seattle that are very cold.
* H3: During the days on which the temperatures are colder, the number of people who commute by bike in the Mountain to Sound Trail West of I90 goes down.

For the second research question RQ2, below are my hypothesis:

* H1: The temperature fluctutations between the colder winter days and the regular warmer days rest of the year would be similar between the two locations, MTS Trail West of I90 and Burke Gilman Trail north of 70th ST.
* H2: The number of people who commute by bike in the Burke Gilman Trail north of 70th St reduces during the colder winter days as compared to the number on other warmer days of the year, similar to the pattern on the MTS Trail west of I90. 

Extending the hypothesis, I think it would be a safe generalization to say that the rest of the trails in the city of Seattle will see similar bike commuter count changes when temperatures go down in winter unless there were any drastic differences between these trails and other trails in the City. 

## Why is this study important?

There are at least two reasons I see why this study is important. 

    1. When people do not commute by bike and take an alternative mode, this behavior leads to more congestion (either directly or indirectly, as elaborated below). More congestion can cause longer commute times and may also contribute to air pollution.

        a. The congestion impact is direct when this alternative mode of commute is car or taxi because it directly adds a motorized vehicle on the road. The side assumption here is that the taxi taken by a person is otherwise free and not on road during that time. Also, the contribution to pollution is only when this motorized vehicle produces exhaust (eg. non-electric vehicles).

        b. The impact is indirect when the alternative mode of commute is a carpool or a bus or a train, because even in this scenario there is an increase in the number of commuting people on the road by means other than bike and are still consuming the transportation resources. In this scenario it can be argued that there is no added impact when these people only increase the occupancy ratio of the existing fleet. And it is only when the fleets exceed their capacity requiring more fleet to be added that the real impact will begin to be noticed.


    2. As we know, bike riding is a very good form of exercise. We could conclude that when people bike to work they are getting better exercise. If these people do not bike to work and assuming some of them do not compensate this exercise with any other form, they are losing out on the health benefit. 

In both the scenarios mentioned above, we can see there is impact to humans and is an example of one of the values of a Human Centered Study Design.

In addition, the results of the study can be used to suggest and drive infrastructure or amenities improvements that can encourage more commuters to ride bikes even during the colder winter days. An example of such an amenities is setting up warm up shelters at a few locations within the trail.

There are a few considerations while accessing these datasets:

1. The Road Weather Information Stations dataset does not contain the temperature informations for the two trail locations chosen for this research. However, there is a closest weather information station to each of these trail locations respectively that can be used as an approximation for the temperature at these trail locations. Below is the closest station selected for each of these trails, each of these is within 2 miles of the corresponding trail and hence is a good approximation assuming that the temperature variation wouldn't be very large for that distance. 

    * For MTS trail, the closest temperature Station Name: JoseRizalBridgeNorth
    * For Burke Gilman trail, the closest temperature station name: NE45StViaduct
    
The details for this are captured in the [ReadMe document for this repository]((https://github.com/sumanbhagavathula/data512-finalprojectreport/blob/master/README.md)
    
2. The bike counter data is collected hourly whereas the road weather information dataset is collected every minute. To be able to join the two datasets, I am going to perform an aggregation of the road weather dataset to average the temperature for an hour. This way the two datasets will have the same granularity for the dates and can be joined seamlessly. The assumption is that the temperature fluctuations within an hour wouldn't be much and hence an average is a good measure to use. 

3. The bike counter data is available starting from 01/01/2014 12:00AM whereas the road weather information data is available from 04/09/2014 09:35AM. I perform an inner join between the two datasets that eliminates the data from the bike counter dataset for the dates that are not available in the road weather information dataset. This should be ok for this research study since the weather information for the remaining dates that are not in the bike counter dataset are anyways not useful. 

4. There seem to be a few duplicates entries in the bike count dataset, so I have used a DISTINCT clause in the Sql query used in the API call (described in the next paragraph) to eliminate the duplicates. This is only eliminating the redundant data and does not cause any data loss.

Now, I define bundle some common code blocks into functions so that they can be reused rather than duplicating the code lines for each execution of the same logic. The generatecalendarmonthdays function is used to generate a list of all dates for the year and month mentioned. The getbikecountinformation function is used to invoke the API endpoint to acquire the bike count information at the required station. The trail location for which the data will be downloaded is defined by the API end point for the dataset. The getroadweatherinformation function is used to acquire the temperature information for the station whose name is passed to the function as an input parameter. Note the use of GROUP BY and avg (for average value) in the Sql query defined in this function. This is in accordance with the consideration #2 in the Datasets Information section which is to overcome the limitation of different date granularities for the different datasets.  

```python
def generatecalendarmonthdays(year,month):
    mycalendar = calendar.Calendar()
    dates=[]
    for date in mycalendar.itermonthdates(year,month):
        yearend=year
        monthend=month+1
        if month == 12:
            yearend=yearend+1
            monthend=1
        if (date >= datetime.date(year, month, 1)) & (date < datetime.date(yearend, monthend, 1)):
            #print date
            dates.append(date.strftime('%Y-%m-%d'))
    return dates
```

```python
def getbikecountinformation(endpointurl,startdate, enddate):
    params = {}
    myquery='SELECT date, bike_north, bike_south WHERE date BETWEEN "' + startdate + 'T00:00:00.000" AND "' +  enddate + 'T23:59:59.000"'
    endpoint = endpointurl+'?$query='+myquery
    api_call = requests.get(endpoint.format(**params))
    try:
        response = api_call.json()
        return(response)
    except:
        print("Unexpected error:", sys.exc_info()[0])
        return
```

```python
def getroadweatherinformation(stationname, startdate, enddate):
    params = {}
    myquery='SELECT avg(airtemperature),date_trunc_ymd(datetime),date_extract_hh(datetime),avg(roadsurfacetemperature) WHERE stationname="'+stationname+'" AND datetime BETWEEN "' + startdate + 'T00:00:00.000" AND "' +  enddate + 'T23:59:59.000" GROUP BY date_trunc_ymd(datetime),date_extract_hh(datetime)'
    endpoint = 'https://data.seattle.gov/resource/ivtm-938t.json?$query='+myquery
    api_call = requests.get(endpoint.format(**params))
    try:
        response = api_call.json()
        return(response)
    except:
        print("Unexpected error:", sys.exc_info()[0])
        return
```
